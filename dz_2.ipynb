### 1.2

```python
dr = []
for i, row in houses.iterrows():
    delta = row['yr_renovated'] - row['yr_built']
    if delta > 0:
        dr.append(delta)
    else:
        dr.append(0)

```

```python
houses['delta_renovated'] = dr
houses.head()
```
